```python
import nibabel as nib
import os
import numpy as np
from utils import preprocess_mask_labels

id_ = 'BraTS-GLI-00707-000'
_whole_slices = nib.load(os.path.join('brats_data', 'val', id_, id_+'-seg.nii.gz'))
_whole_slices = np.asarray(_whole_slices.dataobj) #(240,240,155)
_whole_slices_T = _whole_slices.transpose(2,0,1)
whole_slices = preprocess_mask_labels(_whole_slices_T)
print("after preprocess", whole_slices.shape)
```

```python
import matplotlib.pyplot as plt
plt.subplot(1,2,1)
plt.imshow(_whole_slices[:,:,70])
plt.colorbar()
plt.subplot(1,2,2)
plt.imshow(_whole_slices_T[70])
plt.colorbar()
np.array_equal(_whole_slices[:,:,70], _whole_slices_T[70])
```

```python
wt = whole_slices[0]
tc = whole_slices[1]
et = whole_slices[2]
```

```python
i = 1
start_slice = 68
dt_len = 30
plt.figure(figsize=(3,dt_len))
plt.subplot(dt_len, 3, 1)
plt.title('WT')
plt.subplot(dt_len, 3, 2)
plt.title('TC')
plt.subplot(dt_len, 3, 3)
plt.title('ET')
for j in range(start_slice,start_slice+dt_len):
    plt.subplot(dt_len,3,i)
    plt.imshow(wt[j], cmap='gray')
    plt.axis('off')
    i+=1
    plt.subplot(dt_len,3,i)
    plt.imshow(tc[j], cmap='gray')
    plt.axis('off')
    i+=1
    plt.subplot(dt_len,3,i)
    plt.imshow(et[j], cmap='gray')
    plt.axis('off')
    i+=1
```

```python
def show_slice(slice2show):
    plt.suptitle(f'slice {slice2show}')
    plt.subplot(1,3,1)
    plt.imshow(wt[slice2show],cmap='gray')
    plt.axis('off')
    plt.subplot(1,3,2)
    plt.imshow(tc[slice2show],cmap='gray')
    plt.axis('off')
    plt.subplot(1,3,3)
    plt.imshow(et[slice2show],cmap='gray')
    plt.axis('off')
    plt.tight_layout()
    plt.show()
```

```python
show_slice(58)
show_slice(58+1)
show_slice(58+2)
```
