```python
from unetpp import UNETPP
model_info = {
        "encoder": [
            [16, 32],
            [32, 64],
            [64, 128],
            [128, 256],
        ],
        'in_channel': 3
    }
model = UNETPP(model_info)
model.eval()
```

```python
import torch
img_ex = torch.randn((1,3,64,128,128))
img_ex.shape
```

```python
res = model(img_ex)
res.shape
```
